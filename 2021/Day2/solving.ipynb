```python
import pandas as pd
import numpy as np
```

```python
input = pd.read_csv('input.txt',header=None)
example = pd.read_csv('example.txt', header=None)
```

```python
input['dir']=input.loc[:,0].apply(lambda x: x.split()[0])
input['amount']=input.loc[:,0].apply(lambda x: int(x.split()[1]))
input.head()
```

```python
depth = input.loc[input['dir'] == 'down', 'amount'].sum() - input.loc[input.dir == 'up', 'amount'].sum()
distance = input.loc[input.dir == 'forward', 'amount'].sum()
depth*distance
```

# Part 2

## Set the aim for every forward move

```python
input['aim'] = np.nan
input.loc[input.dir == 'up', 'aim'] = -input.loc[input.dir == 'up', 'amount']
input.loc[input.dir == 'down', 'aim'] = input.loc[input.dir == 'down', 'amount']
input.loc[0,'aim'] = 0
# input.aim = input.aim.ffill()
input['aim_cumsum']=input.aim.cumsum()
# input.aim_cumsum = 
input['aim_cumsum_ffill'] = input.aim_cumsum.fillna(method='ffill')
input.head(15)
```

```python
input['forw'] = input.loc[input.dir=='forward', 'amount']
forward = input.forw.sum()
# input.forw=input.forw.ffill()
input['depth']=input.forw * input.aim_cumsum_ffill
input
```

```python
(forward,input.depth.sum(), forward*input.depth.sum())
```
